```python
x = [1,2,3,4]
y = [12, 14, 19,22]
plt.plot(x,y);
```

```python
# 1st method

fig = plt.figure() #create a figure
ax = fig.add_subplot() # adds some axis
plt.show()
```

```python
# 2nd method

fig = plt.figure() #create a figure
ax = fig.add_axes([1,1,1,1]) # adds some axis
ax.plot(x,y)
plt.show()
```

```python
### 3rd method

fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots()
ax.plot(x,[25, 50, 75, 100]);
```
